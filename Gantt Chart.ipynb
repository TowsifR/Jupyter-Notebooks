#### Date can be written as the year only ('2021'), year and month ('2021-06') or year, month, and day ('2021-08-14')
#### The different roles are PI, Undergraduate Student, Thesis Student, Research Assistant, MSc Student, PhD Student, Technician, Postdoc, and Research Associate
#### After adding members, run both cells to get the updated chart

```python
import plotly.express as px
import pandas as pd
import plotly.io as pio


#Add Users Here

df = pd.DataFrame([
    dict(Name="Mojca Mattiazzi Usaj", Start='2021-01', Finish='2022', Status='PI'),
    dict(Name="Kenneth ", Start='2021-05', Finish='2021-08', Status="Undergraduate Student"),
    dict(Name="Kenneth G. Antenor ", Start='2021-09', Finish='2022-04', Status="Thesis Student"),
    dict(Name="Larry Nguyen", Start='2021-05', Finish='2021-08', Status="Undergraduate Student"),
    dict(Name="Wasiur Rahman", Start='2021-05', Finish='2021-08', Status="Undergraduate Student")
    
])
```

```python

#Sorting by start date
df["Start"] = pd.to_datetime(df["Start"])
df = df.sort_values(by="Start")
df=df.reset_index(drop=True)
df=df.iloc[::-1]


#Creating plot
fig = px.timeline(df, 
                  x_start="Start", 
                  x_end="Finish", 
                  y="Name", 
                  title='Lab Members', 
                  template="simple_white",
                  color="Status",
                  color_discrete_map={
                  #Colour codes for each role
                  "PI": "#f48c06",
                  "Undergraduate Student": "#0096c7",
                  "Thesis Student": "#023e8a",
                  "Research Assistant":"#48cae4",
                   "MSc Student":"#1a7431",
                  "PhD Student": "#25a244",
                   "Technician":"#7b2cbf",
                  "Postdoc": "#c77dff",
                   "Research Associate": "#3c096c"
                  
                  }
                 )

png_renderer = pio.renderers["png"]
png_renderer.width = 900
png_renderer.height = 500
pio.renderers.default = "png"


fig.update_xaxes(title="Date Joined",dtick="M1",tickformat="%b\n%Y",showgrid=True, gridwidth=2.5)
fig.show()
```
